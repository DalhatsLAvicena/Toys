```python
# muat data banjir
import pandas as pd

banjir = pd.read_excel("data_banjir.xlsx", index_col=0)
banjir.head()
```

## Penyusunan Pertanyaan 
1. Berapa jumlah total korban jiwa (meninggal, terluka, dan hilang) dari banjir?
2. Provinsi mana yang memiliki peluang banjir paling tinggi?

### Total Korban Jiwa Akibat Banjir

```python
# Total korban jiwa = hilang + meninggal + terluka
banjir["Korban"] = banjir["Meninggal"] + banjir["Hilang"] + banjir["Terluka"]
# banjir.head() # test
# total korban banjir
korban_banjir = banjir["Korban"].sum()
print("Korban Banjir =", korban_banjir)
```

### Jumlah Kejadian Banjir Berdasarkan Provinsi

```python
# Hitung kejadian banjir berdasarkan provinsi
kejadian = pd.DataFrame(banjir["Provinsi"].value_counts())
jumlah = kejadian["count"].sum()
kejadian["Peluang (%)"] = kejadian["count"]/jumlah*100
kejadian
```

```python
provinsi = kejadian["count"].idxmax()
print("Provinsi paling rentan banjir =", provinsi)
```
